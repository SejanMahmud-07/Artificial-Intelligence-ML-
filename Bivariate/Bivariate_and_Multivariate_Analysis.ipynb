# Today we will learn Bivariate and Multivariate analysis

```python
import pandas as pd
import seaborn as sns
```

# In seaborn there is a BUILTIN DATASET CALLED tips

```python
tips = sns.load_dataset('tips')
tips.head()
```

# we have a dataset called titanic dataset

```python
titanic = pd.read_csv('train.csv')
titanic.head()
```

# In seaborn there is a BUILTIN DATASET CALLED flight

```python
flights = sns.load_dataset('flights')
flights.head()
```

# In seaborn there is a BUILTIN DATASET CALLED iris

```python
iris = sns.load_dataset('iris')
iris
```


# 1. Scatterplot (Numerical - Numerical)

using scatterplot we can see relation between two columns.

for example we will determine the relationship between total_bill and tip

```python
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

Findout male and female customer

```python
sns.scatterplot(x='total_bill', y='tip',
                data=tips, hue='sex')
```

# Multivariate analysis
Here there are relation between 4 columns
Findout male and female customer & who is smoker and who is not

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex',style='smoker')
```

# 2. Bar Plot (Numerical - Categorical)
where there is a two column One is Numerical and one is Categorical and we need to build relation between them , in this case we use Bar plot.



```python
titanic.head()
```

For example we want to determine relation between pclass(Categorical) & Age(Numerical)

```python
sns.barplot(x='Pclass',y='Age',data=titanic)
```

relation between pclass and fair

```python
sns.barplot(x='Pclass',y='Fare',data=titanic)
```

# Multivariate analysis
Relation between pclass,Age,Sex

```python
sns.barplot(x='Pclass',y='Age',hue='Sex',data=titanic)
```

# 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x='Sex',y='Age',hue='Sex',data=titanic)
```

```python
sns.boxplot(x='Sex',y='Age',hue='Survived',data=titanic)
```

# 4. Distplot (Numerical - Categorical)

Person who didnt survived

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
```

Person who  survived with respect to age

```python
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

Relation between survived and not survived with respect to age

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

# 5. HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

Generate heatmap

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

# How many male or female survived

(titanic.groupby('Sex').mean()['Survived']*100)

```python
(titanic.groupby('Sex').mean()['Survived']*100)
```

```python
(titanic.groupby('Embarked').mean()['Survived']*100)
```

# 6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```
