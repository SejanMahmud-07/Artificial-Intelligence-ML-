# we will learn univariate data analysys today

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df.head()
```

# 1.Categorical Data

# countplot
we will determine how many people survibed or not

```python
sns.countplot(df['Survived'])
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
df['Pclass'].value_counts().plot(kind='bar')
```

```python
df['Sex'].value_counts().plot(kind='bar')
```

# b. PieChart
if we want to show percentage of data we should use pie chart

```python
df['Sex'].value_counts().plot(kind='pie')
```

For showing percentage we have to write this code

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Age'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Embarked'].value_counts().plot(kind='pie',autopct='%.2f')
```
